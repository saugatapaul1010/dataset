```python
import time
import os
from tqdm import tqdm
import glob
from PIL import Image
import random
import pandas as pd
```

```python
source = "png"

#Get all the file names in a single list
file_names = []
state_names = os.listdir(source)
for snames in tqdm(state_names):
    worktype_folder = os.listdir(source+"/"+snames)
    for worktype in worktype_folder:
        police_station_name = os.listdir(source+"/"+snames+"/"+worktype)
        for station_name in police_station_name:
            files = os.listdir(source+"/"+snames+"/"+worktype+"/"+station_name)
            for f_name in files:
                f_string = "{}/{}/{}/{}".format(source,snames,worktype,station_name) + "/" + f_name
                file_names.append(f_string)
```

```python
"""
Poor Quality => 1
Blurred => 2
Noisy => 3
High Quality => 4
Has color => 5
Handwritten => 6
"""

source = "Non-Handwritten-California/*"
image_list = glob.glob(source)

labels = []
filenames = random.sample(file_names,4)
for image in filenames:
    im = Image.open(image)
    im.show()
    label = int(input("Enter the image label: "))
    labels.append(label)
    im.close()
    time.sleep(1)
    
def process_names(text):
    file_name = text.split("/")[1]
    return file_name

df = pd.DataFrame(columns=["image_name","numeric_labels"])
df["image_name"]=filenames
df["numeric_labels"]=labels
df["image_name"]=df["image_name"].apply(lambda x: process_names(x))

df.to_csv("image_labels.csv")
df.head()
```

```python
def process_int(digit):
    if(digit==1):
        return "Poor Quality"
    elif(digit==2):
        return "Blurred"
    elif(digit==3):
        return "Noisy"
    elif(digit==4):
        return "Good Quality"
    elif(digit==5):
        return "Has Color"
    elif(digit==6):
        return "Handwritten"
```

```python
df["quality_labels"]=df["labels"].apply(lambda x: process_int(x))
df.head()
```
